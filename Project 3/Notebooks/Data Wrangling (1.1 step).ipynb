```python
filtered_df = data_df[data_df['Indicator Code'].isin(indicators)]

# Reshape the data from wide to long format
long_df = pd.melt(filtered_df, id_vars=['Country Code','Indicator Code'], var_name='Year', value_name='Value')

# Convert the 'Year' column to datetime format for time series handling
long_df['Year'] = pd.to_datetime(long_df['Year'], format='%Y')
display(long_df.head())
long_df.to_csv('../data/processed/long_df.csv')
```

```python
data = long_df.pivot_table(
    index=['Country Code', 'Year'],
    columns='Indicator Code',
    values='Value'
).reset_index()

data.columns.name = None

display(data.head())
```
