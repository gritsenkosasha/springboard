 Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
cursor.execute("select f.name as facility_name,sum(slots*(case when b.memid = 0 then guestcost ELSE membercost end)) as revenue from Bookings b left join Facilities f on f.facid = b.facid GROUP by f.name HAVING revenue < 1000 ORDER by revenue ;")
r = cursor.fetchall()

# Extract the table names from the result
result = [row for row in r]

result
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */


```python
cursor.execute("SELECT mc.surname||' '||mc.firstname as Member, mr.surname||' '||mr.firstname as Recommended_by FROM Members mc LEFT JOIN Members mr ON mc.recommendedby = mr.memid WHERE LENGTH(mc.recommendedby) > 0 ORDER BY Member ASC;")
r11 = cursor.fetchall()
df11 = pd.DataFrame(r11)
```

```python
df11.columns = [i[0] for i in cursor.description]
display(df11)
```

/* Q12: Find the facilities with their usage by member, but not guests */

```python
cursor.execute("SELECT Name as Facility , count(bookid) as Usage FROM Bookings b left join Facilities f on f.facid=b.facid where memid <>0 GROUP by Name Order by usage ;")
r12 = cursor.fetchall()
df12 = pd.DataFrame(r12)

df12.columns = [i[0] for i in cursor.description]
display(df12)
```

/* Q13: Find the facilities usage by month, but not guests */

```python
cursor.execute("SELECT Name as Facility, strftime('%m', starttime) as Month, count(bookid) as Usage FROM Bookings b LEFT JOIN Facilities f ON f.facid = b.facid WHERE memid <> 0 GROUP BY month, Facility;")
r13 = cursor.fetchall()
df13 = pd.DataFrame(r13)

df13.columns = [i[0] for i in cursor.description]
display(df13)
```
